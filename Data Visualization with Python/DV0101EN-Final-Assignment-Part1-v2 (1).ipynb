````python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import requests
import io

# URL of the CSV file
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Fetch the data from the URL
response = requests.get(URL)

# Raise an error if the request failed
response.raise_for_status()

# Convert the response content into a readable format for pandas
csv_content = io.StringIO(response.text)

# Read the CSV data into a pandas dataframe
df = pd.read_csv(csv_content)

print('Data downloaded and read into a dataframe!')

# Create data for plotting
df_line = df.groupby(df['Year'])['Automobile_Sales'].mean()

# Create figure
plt.figure(figsize=(10, 6))
df_line.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Automobile Sales during Recession')
plt.xticks(list(range(1980, 2024)), rotation=75)

# Annotate recession years
plt.text(1981, df_line.loc[1981] * 1.05, '1981-82 Recession', rotation=45, color='red')
plt.text(2008, df_line.loc[2008] * 1.05, '2008-09 Recession', rotation=45, color='red')
plt.text(2020, df_line.loc[2020] * 1.05, '2020 Covid-19', rotation=45, color='red')

plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import io

# Data loading (as provided in the notebook)
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
response = requests.get(URL)
response.raise_for_status()
csv_content = io.StringIO(response.text)
df = pd.read_csv(csv_content)

# Create data for plotting: group by year and calculate mean Automobile_Sales
df_line = df.groupby(df['Year'])['Automobile_Sales'].mean()

# Create figure
plt.figure(figsize=(10, 6))

# Plot the line chart
df_line.plot(kind='line')

# Add x-axis label, y-axis label, and title
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Automobile Sales during Recession')

# Set x-axis ticks to show all years and rotate them for readability
plt.xticks(list(range(1980, 2024)), rotation=75)

# Add annotations for at least two years of recession
# Recession periods from the Data Description:
# recession period 1 - year 1980
# recession period 2 - year 1981 to 1982
# recession period 3 - year 1991
# recession period 4 - year 2000 to 2001
# recession period 5 - year end 2007 to mid 2009
# recession period 6 - year 2020 -Feb to April (Covid-19 Impact)

plt.text(1981, df_line.loc[1981] * 1.05, '1981-82 Recession', rotation=45, color='red', ha='center')
plt.text(2008, df_line.loc[2008] * 1.05, '2008-09 Recession', rotation=45, color='red', ha='center')
plt.text(2020, df_line.loc[2020] * 1.05, '2020 Covid-19', rotation=45, color='red', ha='center')


plt.grid(True, linestyle='--', alpha=0.6) # Add a grid for better readability
plt.legend(['Average Sales']) # Add a legend
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
````

````python
# Filter data for recession periods
df_rec = df[df['Recession'] == 1]

# Group by Year and Vehicle_Type to calculate mean Automobile_Sales
df_Mline = df_rec.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()

# Set 'Year' as index and then group by 'Vehicle_Type' for plotting multiple lines
df_Mline.set_index('Year', inplace=True)
df_Mline = df_Mline.groupby(['Vehicle_Type'])['Automobile_Sales']

# Plotting
plt.figure(figsize=(12, 7))
df_Mline.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Sales Trend Vehicle-wise during Recession')
plt.legend(title='Vehicle Type')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
# Create dataframes for recession and non-recession period
rec_data = df[df['Recession'] == 1]
non_rec_data = df[df['Recession'] == 0]

# Create figure
fig = plt.figure(figsize=(15, 6))

# Create different axes for subplotting
ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot)

# Plot for Recession Period
sns.lineplot(x='Year', y='GDP', data=rec_data, label='Recession', ax=ax0, color='red')
ax0.set_xlabel('Year')
ax0.set_ylabel('GDP')
ax0.set_title('GDP Variation during Recession Period')
ax0.grid(True, linestyle='--', alpha=0.6)
ax0.legend()

# Plot for Non-Recession Period
sns.lineplot(x='Year', y='GDP', data=non_rec_data, label='Non-Recession', ax=ax1, color='green')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')
ax1.set_title('GDP Variation during Non-Recession Period')
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.legend()

plt.tight_layout()
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
non_rec_data = df[df['Recession'] == 0]

# Use Seasonality_Weight for bubble size
size = non_rec_data['Seasonality_Weight'] * 500 # Multiply for better visibility of bubble size

plt.figure(figsize=(12, 7))
sns.scatterplot(data=non_rec_data, x='Month', y='Automobile_Sales', size=size, hue='Seasonality_Weight', legend='full', sizes=(50, 1000))
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.title('Seasonality impact on Automobile Sales (Non-Recession Periods)')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
````

````python
# Filter the data for recession period
Rdata = df[df['Recession'] == 1]

# Calculate the sum of advertising expenditure by vehicle type during recessions
VTexpenditure = Rdata.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()

# Create a pie chart for the share of each vehicle type in total expenditure during recessions
plt.figure(figsize=(10, 10))

labels = VTexpenditure.index
sizes = VTexpenditure.values
colors = sns.color_palette('pastel')[0:len(labels)] # Get a color palette

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, pctdistance=0.85)
# Add a circle at the center to make it a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Share of Each Vehicle Type in Total Advertising Expenditure during Recessions')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
import io

# Data loading (as provided in the notebook)
# (Assuming this part runs successfully now that you've confirmed it works)
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
response = requests.get(URL)
response.raise_for_status()
csv_content = io.StringIO(response.text)
df = pd.read_csv(csv_content)

print('Data downloaded and read into a dataframe!')

# Filter the data for the recession period
df_rec = df[df['Recession'] == 1]

# TASK 1.9: Develop a lineplot to analyse the effect of the unemployment rate on vehicle type and sales during the Recession Period.
# Corrected column name: 'unemployment_rate' instead of 'Unemployment_Rate'
plt.figure(figsize=(12, 7))
sns.lineplot(data=df_rec, x='unemployment_rate', y='Automobile_Sales',
             hue='Vehicle_Type', style='Vehicle_Type', markers=True, err_style=None)
plt.ylim(0, 850) # Set y-axis limit for better visualization
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1), title='Vehicle Type') # Adjust legend position
plt.xlabel('Unemployment Rate') # Label can still be capitalized for readability
plt.ylabel('Automobile Sales')
plt.title('Effect of Unemployment Rate on Vehicle Type and Sales during Recession Period')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
````

````python
import folium
import pandas as pd
import requests
import io
import os # Import os module to check for file existence

# Re-load data just in case, though it should be loaded from previous tasks
URL_CSV = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
response_csv = requests.get(URL_CSV)
response_csv.raise_for_status()
csv_content = io.StringIO(response_csv.text)
df = pd.read_csv(csv_content)

print('Main data downloaded and read into a dataframe!')

# --- Start of handling us-states.json ---
geojson_filename = "us-states.json"
URL_GEOJSON = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/us-states.json'

# Check if the geojson file already exists
if not os.path.exists(geojson_filename):
    print(f"Downloading {geojson_filename}...")
    try:
        response_geojson = requests.get(URL_GEOJSON)
        response_geojson.raise_for_status() # Raise an exception for HTTP errors
        with open(geojson_filename, "wb") as f:
            f.write(response_geojson.content)
        print(f"{geojson_filename} downloaded successfully!")
    except requests.exceptions.RequestException as e:
        print(f"Error downloading {geojson_filename}: {e}")
        geojson_filename = None # Set to None if download fails to prevent further errors
else:
    print(f"{geojson_filename} already exists locally.")

# --- End of handling us-states.json ---


# Filter the data for the recession period
recession_data = df[df['Recession'] == 1]

# Calculate the total sales by city
# The 'City' column in the dataset actually represents US states based on its values.
# Let's assume 'City' refers to states for the purpose of using us-states.json.
sales_by_state = recession_data.groupby('City')['Automobile_Sales'].sum().reset_index()

# Rename 'City' to 'name' to match the 'feature.properties.name' in the GeoJSON for mapping
sales_by_state.rename(columns={'City': 'name'}, inplace=True)

# Create a base map centered on the United States
map1 = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

if geojson_filename and os.path.exists(geojson_filename):
    # Create a choropleth layer using Folium
    choropleth = folium.Choropleth(
        geo_data=geojson_filename,  # Path to the GeoJSON file
        data=sales_by_state,
        columns=['name', 'Automobile_Sales'],
        key_on='feature.properties.name', # Key in the GeoJSON that matches our 'name' column
        fill_color='YlOrRd', # Color scheme (Yellow-Orange-Red)
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name='Automobile Sales during Recession (USD)'
    ).add_to(map1) # Add to the map

    # Add tooltips to the choropleth layer for interactivity
    choropleth.geojson.add_child(
        folium.features.GeoJsonTooltip(['name'], labels=True) # Show state name on hover
    )

    # Display the map (this will show the map in your Jupyter output)
    print("Displaying the choropleth map...")
    map1
else:
    print("Map cannot be displayed because the GeoJSON file is not available.")
````

````python
import folium
import pandas as pd
import requests
import io

# Re-load data just in case, though it should be loaded from previous tasks
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
response = requests.get(URL)
response.raise_for_status()
csv_content = io.StringIO(response.text)
df = pd.read_csv(csv_content)

# Filter the data for the recession period
recession_data = df[df['Recession'] == 1]

# Calculate the total sales by city
# The 'City' column in the dataset actually represents US states based on its values.
# Let's assume 'City' refers to states for the purpose of using us-states.json.
sales_by_city = recession_data.groupby('City')['Automobile_Sales'].sum().reset_index()

# Rename 'City' to 'name' to match the 'feature.properties.name' in the GeoJSON for mapping
sales_by_city.rename(columns={'City': 'name'}, inplace=True)

# Create a base map centered on the United States
map1 = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# Load the GeoJSON file (assuming 'us-states.json' is available locally)
# For this step to work, 'us-states.json' must be in the same directory or accessible via a path.
try:
    with open("us-states.json", 'r') as f:
        geojson_data = f.read()
except FileNotFoundError:
    print("Error: us-states.json not found. Please ensure it's in the same directory or downloaded.")
    geojson_data = None # Set to None to prevent further errors

if geojson_data:
    # Create a choropleth layer using Folium
    choropleth = folium.Choropleth(
        geo_data=geojson_data,  # GeoJSON file with state boundaries
        data=sales_by_city,
        columns=['name', 'Automobile_Sales'],
        key_on='feature.properties.name', # Key in the GeoJSON that matches our 'name' column
        fill_color='YlOrRd', # Color scheme
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name='Automobile Sales during Recession (USD)'
    ).add_to(map1) # Add to the map

    # Add tooltips to the choropleth layer for interactivity
    choropleth.geojson.add_child(
        folium.features.GeoJsonTooltip(['name'], labels=True) # Show state name on hover
    )

    # Display the map (this will show the map in your Jupyter output)
    map1
else:
    print("Map cannot be displayed due to missing GeoJSON data.")
````
