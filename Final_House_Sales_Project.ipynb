```python
# Q0: Load Dataset
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline

df = pd.read_csv('D:\kc_house_data.csv')
df.head()
```

```python
# Q1: Display the data types of each column using .dtypes
df.dtypes
```

```python
# Q2: Drop 'id' and 'Unnamed: 0' columns and use describe()
df.drop(['id'], axis=1, inplace=True)
df.describe()
```

```python
# Q3: Use value_counts on 'floors' and convert to DataFrame
df['floors'].value_counts().to_frame()
```

```python
# Q4: Use seaborn boxplot to compare price with/without waterfront view
sns.boxplot(x='waterfront', y='price', data=df)
plt.show()
```

```python
# Q5: Use seaborn regplot to show correlation between sqft_above and price
sns.regplot(x='sqft_above', y='price', data=df)
plt.show()
```

```python
# Q6: Linear Regression with 'sqft_living' as predictor
lm = LinearRegression()
X = df[['sqft_living']]
y = df['price']
lm.fit(X, y)
lm.score(X, y)
```

```python
# Q7: Linear Regression with multiple predictors
features = ['floors', 'waterfront', 'lat', 'bedrooms', 'sqft_basement', 'view',
            'bathrooms', 'sqft_living15', 'sqft_above', 'grade', 'sqft_living']
X = df[features]
y = df['price']
lm.fit(X, y)
lm.score(X, y)
```

```python
# Q8: Pipeline with scaling, polynomial transform, and linear regression
Input = df[features]
y = df['price']
pipe = Pipeline([
    ('scale', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2)),
    ('model', LinearRegression())
])
pipe.fit(Input, y)
pipe.score(Input, y)
```

```python
# Q9: Ridge regression (alpha=0.1) on training data
X_train, X_test, y_train, y_test = train_test_split(Input, y, test_size=0.1, random_state=1)
ridge = Ridge(alpha=0.1)
ridge.fit(X_train, y_train)
ridge.score(X_test, y_test)
```

```python
# Q10: Ridge regression with polynomial features (degree=2)
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
ridge_poly = Ridge(alpha=0.1)
ridge_poly.fit(X_train_poly, y_train)
ridge_poly.score(X_test_poly, y_test)
```
