```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import itertools

# Load the dataset 
file_path = r"C:\Users\maste\Downloads\IBM Data Analyst Capstone Project - module 6\survey_data_updated 5 (1).csv"
df = pd.read_csv(file_path)

# Helper function to get top N counts from semicolon-separated string entries
def get_top_counts(series, top_n=10):
    all_items = series.dropna().apply(lambda x: x.split(';'))
    flat_list = list(itertools.chain.from_iterable(all_items))
    counts = Counter(flat_list)
    return counts.most_common(top_n)

# Get top 10 programming languages (current and future)
top_current_langs = get_top_counts(df["LanguageHaveWorkedWith"])
top_future_langs = get_top_counts(df["LanguageWantToWorkWith"])

# Convert to DataFrames
df_current_langs = pd.DataFrame(top_current_langs, columns=["Language", "Count"])
df_future_langs = pd.DataFrame(top_future_langs, columns=["Language", "Count"])

# Plot 1: Top 10 Programming Languages Currently Used
plt.figure(figsize=(10, 6))
sns.barplot(data=df_current_langs, x="Count", y="Language", palette="Blues_d")
plt.title("Top 10 Programming Languages Currently Used")
plt.xlabel("Number of Mentions")
plt.ylabel("Programming Language")
plt.tight_layout()
plt.savefig("top_current_languages.png")  # Saves chart as file
plt.show()

# Plot 2: Top 10 Programming Languages Wanted in the Future
plt.figure(figsize=(10, 6))
sns.barplot(data=df_future_langs, x="Count", y="Language", palette="Greens_d")
plt.title("Top 10 Programming Languages Wanted in the Future")
plt.xlabel("Number of Mentions")
plt.ylabel("Programming Language")
plt.tight_layout()
plt.savefig("top_future_languages.png")  # Saves chart as file
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import itertools

# Load the survey dataset
file_path = r"C:\Users\maste\Downloads\IBM Data Analyst Capstone Project - module 6\survey_data_updated 5 (1).csv"



# Load dataset
df = pd.read_csv(file_path)

# Helper for top N from semicolon-separated fields
def get_top_counts(series, top_n=10):
    items = series.dropna().apply(lambda x: x.split(";"))
    flat = list(itertools.chain.from_iterable(items))
    return pd.DataFrame(Counter(flat).most_common(top_n), columns=["Item", "Count"])

# ---------- Dashboard 1: Current Technology Usage ----------
lang = get_top_counts(df["LanguageHaveWorkedWith"])
db = get_top_counts(df["DatabaseHaveWorkedWith"])
web = get_top_counts(df["WebframeHaveWorkedWith"])
os = df["OpSysPersonal use"].value_counts().nlargest(6)

fig, axs = plt.subplots(2, 2, figsize=(18, 12))
sns.barplot(data=lang, y="Item", x="Count", ax=axs[0,0], palette="Blues_d")
axs[0,0].set_title("Languages Used")

sns.barplot(data=db, y="Item", x="Count", ax=axs[0,1], palette="Purples_d")
axs[0,1].set_title("Databases Used")

sns.barplot(data=web, y="Item", x="Count", ax=axs[1,0], palette="Greys_d")
axs[1,0].set_title("Web Frameworks Used")

sns.barplot(x=os.values, y=os.index, ax=axs[1,1], palette="Reds_d")
axs[1,1].set_title("OS (Personal Use)")

plt.suptitle("Dashboard 1: Current Technology Usage", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("dashboard_current_tech_extended.png")
plt.show()

# ---------- Dashboard 2: Future Technology Trends ----------
lang_future = get_top_counts(df["LanguageWantToWorkWith"])
db_future = get_top_counts(df["DatabaseWantToWorkWith"])
ai_interested = get_top_counts(df["AIToolInterested in Using"])
cloud = get_top_counts(df["ProfessionalCloud"])

fig, axs = plt.subplots(2, 2, figsize=(18, 12))
sns.barplot(data=lang_future, y="Item", x="Count", ax=axs[0,0], palette="Greens_d")
axs[0,0].set_title("Languages Wanted")

sns.barplot(data=db_future, y="Item", x="Count", ax=axs[0,1], palette="Oranges_d")
axs[0,1].set_title("Databases Wanted")

sns.barplot(data=ai_interested, y="Item", x="Count", ax=axs[1,0], palette="YlGnBu")
axs[1,0].set_title("AI Tools Interested In")

sns.barplot(data=cloud, y="Item", x="Count", ax=axs[1,1], palette="coolwarm")
axs[1,1].set_title("Cloud Platforms Used Professionally")

plt.suptitle("Dashboard 2: Future Technology Trends", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("dashboard_future_tech_extended.png")
plt.show()

# ---------- Dashboard 3: Developer Demographics ----------
age = df["Age"].value_counts().sort_index()
edu = df["EdLevel"].value_counts().nlargest(6)
emp = df["Employment"].value_counts().nlargest(6)
devtype = get_top_counts(df["DevType"])
years_code = df["YearsCode"].value_counts().sort_index().drop(["Less than 1 year", "More than 50 years"], errors='ignore').nlargest(10)

fig, axs = plt.subplots(2, 2, figsize=(18, 12))
sns.barplot(x=age.values, y=age.index, ax=axs[0,0], palette="pastel")
axs[0,0].set_title("Age Distribution")

sns.barplot(x=edu.values, y=edu.index, ax=axs[0,1], palette="Set2")
axs[0,1].set_title("Education Level")

sns.barplot(x=emp.values, y=emp.index, ax=axs[1,0], palette="Set3")
axs[1,0].set_title("Employment Type")

sns.barplot(data=devtype.head(10), y="Item", x="Count", ax=axs[1,1], palette="BuPu")
axs[1,1].set_title("Top Developer Roles")

plt.suptitle("Dashboard 3: Developer Demographics", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("dashboard_demographics_extended.png")
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import itertools

# Load dataset
file_path = r"C:\Users\maste\Downloads\IBM Data Analyst Capstone Project - module 6\survey_data_updated 5 (1).csv"
df = pd.read_csv(file_path)

# Helper function
def get_top_counts(series, top_n=10):
    items = series.dropna().apply(lambda x: x.split(";"))
    flat = list(itertools.chain.from_iterable(items))
    return pd.DataFrame(Counter(flat).most_common(top_n), columns=["Item", "Count"])

# ---------- Dashboard 1: Current Technology Usage ----------
lang = get_top_counts(df["LanguageHaveWorkedWith"])
db = get_top_counts(df["DatabaseHaveWorkedWith"])
web = get_top_counts(df["WebframeHaveWorkedWith"])
os = df["OpSysPersonal use"].value_counts().nlargest(5)

fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# Lang: Vertical bar
sns.barplot(data=lang, x="Item", y="Count", ax=axs[0,0], palette="Blues_d")
axs[0,0].set_title("Languages Used")
axs[0,0].tick_params(axis='x', rotation=45)

# DB: Pie chart
axs[0,1].pie(db["Count"], labels=db["Item"], autopct="%1.1f%%", startangle=140)
axs[0,1].set_title("Database Usage (Pie Chart)")

# Web Framework: Horizontal bar
sns.barplot(data=web, y="Item", x="Count", ax=axs[1,0], palette="Purples")
axs[1,0].set_title("Web Frameworks Used")

# OS: Count plot (from Series)
sns.countplot(y=df["OpSysPersonal use"], order=df["OpSysPersonal use"].value_counts().index[:5], ax=axs[1,1], palette="coolwarm")
axs[1,1].set_title("Top 5 Personal OS")

plt.suptitle("Dashboard 1: Current Technology Usage", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("dashboard_current_mixed.png")
plt.show()

# ---------- Dashboard 2: Future Technology Trends ----------
lang_f = get_top_counts(df["LanguageWantToWorkWith"])
db_f = get_top_counts(df["DatabaseWantToWorkWith"])
ai = get_top_counts(df["AIToolInterested in Using"])
cloud = get_top_counts(df["ProfessionalCloud"])

fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# Lang: Horizontal bar
sns.barplot(data=lang_f, y="Item", x="Count", ax=axs[0,0], palette="Greens_d")
axs[0,0].set_title("Languages Wanted")

# DB: Donut chart
wedges, texts, autotexts = axs[0,1].pie(db_f["Count"], labels=db_f["Item"], autopct="%1.1f%%", startangle=90)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
axs[0,1].add_artist(centre_circle)
axs[0,1].set_title("Databases Wanted (Donut)")

# AI Tools: Vertical bar
sns.barplot(data=ai, x="Item", y="Count", ax=axs[1,0], palette="YlGnBu")
axs[1,0].set_title("AI Tools (Interest)")
axs[1,0].tick_params(axis='x', rotation=45)

# Cloud: Bubble chart
cloud["Size"] = cloud["Count"] / 100  # scale for bubble size
axs[1,1].scatter(cloud.index, [1]*len(cloud), s=cloud["Size"], alpha=0.6)
axs[1,1].set_yticks([])
axs[1,1].set_xticks(range(len(cloud)))
axs[1,1].set_xticklabels(cloud["Item"], rotation=45)
axs[1,1].set_title("Cloud Platform Bubble Chart")

plt.suptitle("Dashboard 2: Future Technology Trends", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("dashboard_future_mixed.png")
plt.show()

# ---------- Dashboard 3: Developer Demographics ----------
age = df["Age"].value_counts().sort_index()
edu = df["EdLevel"].value_counts().nlargest(5)
emp = df["Employment"].value_counts().nlargest(5)
devtype = get_top_counts(df["DevType"])

fig, axs = plt.subplots(2, 2, figsize=(16, 12))

# Age: Pie chart
axs[0,0].pie(age.values, labels=age.index, autopct="%1.1f%%", startangle=90)
axs[0,0].set_title("Age Distribution")

# Education: Vertical bar
sns.barplot(x=edu.values, y=edu.index, ax=axs[0,1], palette="Set2")
axs[0,1].set_title("Education Level")

# Employment: Horizontal bar
sns.barplot(x=emp.values, y=emp.index, ax=axs[1,0], palette="Set3")
axs[1,0].set_title("Employment Type")

# Dev Type: Count plot
sns.barplot(data=devtype.head(10), y="Item", x="Count", ax=axs[1,1], palette="BuPu")
axs[1,1].set_title("Top Dev Roles")

plt.suptitle("Dashboard 3: Developer Demographics", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("dashboard_demo_mixed.png")
plt.show()
```
